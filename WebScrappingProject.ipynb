```python
root="https://www.football-data.co.uk/mmz4281/"
```

```python
for league in leagues:
    for season in range(23,26):
        df=pd.read_csv(root + str(season)+str((season+1))+"/"+league+".csv")
        df.insert(1, 'season', season)
        frames.append(df)
```

# Organize all data in dict

```python
dict_league={"premier":"E0","Championship":"E1"}
```

```python
dict_historical_data = {} 
for league in dict_league:
    frames=[]
    for season in range(23,26):
        df=pd.read_csv(root + str(season)+str((season+1))+"/"+dict_league[league]+".csv")
        df.insert(1, 'season', season)
        frames.append(df)
        df_frames = pd.concat(frames,ignore_index=True) # storing all seasons of a league in a single dataframe
        dict_historical_data[league] = df_frames # assigning dataframe to a new dictionary
```
